```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv(r"C:\Users\a\Downloads\ml pratice\netfilx\netflix_titles.csv")
```

```python
df
```

```python
df.isnull().mean()*100
```

```python
df['director'] = df['director'].fillna('unknown')
df['director'].isnull().sum()
```

```python
df['cast'] = df['cast'].fillna('unknown')
df['cast'].isnull().sum()
```

```python
df['country'] = df['country'].fillna('unknown')
df['country'].isnull().sum()
```

```python
df.dropna(inplace = True)
df.isnull().sum()
```

```python
df
```

```python
df['country'] = df['country'].apply(lambda x:x.capitalize())
df['country'].head()
```

```python
df[['date_added','release_year']]
```

```python
#df['month_added'] = df['date_added'].apply(lambda x:x.split(',')[0].split()[0])
#df['year_added'] = df['date_added'].apply(lambda x:x.split(',')[1])
#df[['month_added','year_added']]
```

```python
df['month_added'] = df['date_added'].str.split(',').str.get(0).str.split().str.get(0)
df['year_added'] = df['date_added'].str.split(',').str.get(1)
df
```

```python
df.drop('date_added',axis = 1 ,inplace = True)
df
```
